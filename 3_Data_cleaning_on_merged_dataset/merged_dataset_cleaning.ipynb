```python
df_female_employment_clean = pd.read_csv("female_employment_clean.csv")
display(df_female_employment_clean)
```

```python
display(merged_dataset_to_clean)
```

```python
merged_dataset_with_clean_female_empl = merged_dataset_to_clean.merge(
                                                df_female_employment_clean,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=["Country Name", "Country Code", "Year"],
                                                how="inner"
                                                )
display(merged_dataset_with_clean_female_empl)
```

```python
merged_dataset_with_clean_female_empl["Female_employment_rate_x"].isna().sum()
```

```python
merged_dataset_with_clean_female_empl = merged_dataset_with_clean_female_empl.drop(["Female_employment_rate_x"], axis=1)
display(merged_dataset_with_clean_female_empl)
```

```python
merged_dataset_with_clean_female_empl.rename(columns={"Female_employment_rate_y":"Female_employment_rate"}, inplace=True)
display(merged_dataset_with_clean_female_empl)
```

```python
df_GDP_cleaned = pd.read_csv("df_GDP_cleaned.csv", index_col=0)
display(df_GDP_cleaned)
```

```python
df_GDP_cleaned.isna().sum()
```

```python
df_GDP_cleaned = df_GDP_cleaned.drop(df_GDP_cleaned[df_GDP_cleaned['Country Name'] == 'South Sudan'].index)
df_GDP_cleaned = df_GDP_cleaned.drop(df_GDP_cleaned[df_GDP_cleaned['Country Name'] == 'Eritrea'].index)
display(df_GDP_cleaned)
```

```python
df_GDP_cleaned.isna().sum()
```

```python
merged_dataset_with_clean_female_empl.rename(columns={"GDP per Capita":"Old GDP per Capita"}, inplace=True)
display(merged_dataset_with_clean_female_empl)
```

```python
merged_dataset_with_clean_female_empl_and_gdp = merged_dataset_with_clean_female_empl.merge(
                                                df_GDP_cleaned,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=["Country Name", "Country Code", "Year"],
                                                how="inner"
                                                )
display(merged_dataset_with_clean_female_empl_and_gdp)
```

```python
merged_dataset_with_clean_female_empl_and_gdp = merged_dataset_with_clean_female_empl_and_gdp.drop(["Old GDP per Capita"], axis=1)
display(merged_dataset_with_clean_female_empl_and_gdp)
```

```python
merged_dataset_with_clean_female_empl_and_gdp.isna().sum()
```
