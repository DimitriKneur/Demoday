```python
import pandas as pd
```

```python
df_birth_rate_PIB = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/Birth_Rate_PIB.csv")
display(df_birth_rate_PIB)

df_demography = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/Demography_dataset.csv")
display(df_demography)

df_idh = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/IDH_table.csv")
display(df_idh)

df_pollution = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/pollution.csv')
display(df_pollution)

df_education = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/education.csv")
display(df_education)

df_urban = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/urban.csv")
display(df_urban)
```

```python
# Rename column dataset education "value" => "women_education_rate_after_high_school"
df_education.rename(columns={
                             "Value":"Women_education_rate",
                             "COUNTRY":"Country Code",
                             "Country":"Country Name",
                             "YEAR":"Year"
                             }, inplace=True
                    )
display(df_education)
```

```python
merged_final_dataset = df_birth_rate_PIB.merge(
                                                df_demography,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=["Country Name", "Country Code", "Year"],
                                                how="inner"
                                                ).merge(
                                                        df_idh,
                                                        left_on=["Country Name", "Country Code", "Year"],
                                                        right_on=["Entity", "Code", "Year"],
                                                        how="inner"
                                                        ).merge(
                                                                df_pollution,
                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                how="inner"
                                                                ).merge(
                                                                        df_education,
                                                                        left_on=["Country Name", "Country Code", "Year"],
                                                                        right_on=["Country Name", "Country Code", "Year"],
                                                                        how="left"
                                                                        ).merge(
                                                                                df_urban,
                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                how="inner"
                                                                                )
display(merged_final_dataset)
```
